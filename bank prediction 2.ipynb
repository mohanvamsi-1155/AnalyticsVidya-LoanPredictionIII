# Categorical Conversion

```python
data.info()
```

```python
gender=pd.get_dummies(data['Gender'],drop_first=True)
married=pd.get_dummies(data['Married'],drop_first=True)
dependents=pd.get_dummies(data['Dependents'],drop_first=True)
edu=pd.get_dummies(data['Education'],drop_first=True)
self=pd.get_dummies(data['Self_Employed'],drop_first=True,columns=['Self_emp'])
prop=pd.get_dummies(data['Property_Area'],drop_first=True)
status=pd.get_dummies(data['Loan_Status'],drop_first=True)
data.drop(['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status'],axis=1,inplace=True)
data=pd.concat([data,gender,married,dependents,edu,prop,status],axis=1)
```

```python
data.info()
```

```python
self.columns=['self']
```

```python
data=pd.concat([data,self],axis=1)
```

```python
data.info()
```

```python
ses.heatmap(data.corr(),cmap='coolwarm')
```

```python
from xgboost.sklearn import XGBClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data.drop('Y',axis=1),data['Y'],test_size=0.30,random_state=3)
```

```python
rf=XGBClassifier(learning_rate =0.05,
 n_estimators=20,
 max_depth=2,
 min_child_weight=1,
 gamma=0.0,
 subsample=0.8,
 colsample_bytree=0.8,
 objective= 'binary:logistic',
 nthread=5,
 scale_pos_weight=1,
 reg_alpha=1e-5,
 seed=96)
rf.fit(X_train,y_train)
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
parameters = {
                'random_state':[i for i in range(100,200,10)]
             }
xgb_grid = GridSearchCV(rf,
                        parameters,
                        cv = 3,
                        n_jobs = 5,
                        verbose=True)

xgb_grid.fit(X_train,
         y_train)
print(xgb_grid.grid_scores_)
print(xgb_grid.best_score_)
print(xgb_grid.best_params_)
```

```python
predict=rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,predict))
```

```python
print(accuracy_score(y_test,predict))
```

```python
print(precision_score(y_test,predict))
```
